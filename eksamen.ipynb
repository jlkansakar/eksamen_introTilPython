### Python Datamodel


```python
class BookCollection:
    def __init__(self, books):
        self.books = books

    # def __len__(self):
    #     return len(self.books)

    def __str__(self):
        return f"BookCollection with {len(self)} books"

    def __getitem__(self, index):
        return self.books[index]

    def __add__(self, other):
        return BookCollection(self.books + other.books)
```

```python
collection1 = BookCollection(["1984", "Brave New World"])
collection2 = BookCollection(["Dune"])

# print(collection1)                # kalder __str__
print(collection1[0])              # kalder __getitem__


combined = collection1 + collection2  # kalder __add__
# print(combined)
```
