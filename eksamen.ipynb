```python
class BookCollection:
    def __init__(self, books=None):
        if books is None:
            self.books = []
        else:
            self.books = books

    def __len__(self):
        return len(self.books)
    
    def __str__(self):
        return f"BookCollection with {len(self)} books, being ({self.books})"
    
    def __add__(self, other):
        return BookCollection(self.books + other.books)

    def __getitem__(self, key):
        return self.books[key]
    
    def __iter__(self):
        return iter(self.books)
```

```python
# kalder __init__
collection1 = BookCollection(["1984", "Brave New World"])   
collection2 = BookCollection(["Dune"])                      
collection3 = BookCollection([])                            

# kalder __len__
print(len(collection1))

# kalder __str__
print(collection1)                                          
print(collection3)                                          

# kalder __add__
combined = collection1 + collection2 + collection3          

# kalder __str__
print(type(combined))       
print(combined)   

# kalder __getitem__
print(collection1[0]) 

# kalder __iter__                                   
for book in collection1:                                    
    print("the book", book)

# kalder __iter__ 
for book in reversed(collection1):
    print(book, "the book")
    
```
