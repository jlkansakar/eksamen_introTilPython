```python
class BookCollection:
    def __init__(self, books=None):
        # I tilfælde af ingen bøger -> laver collection af 0 bøger
        # # if books is None:
        #     self.books = []
        # else:
        self.books = books

    def __len__(self):
        return len(self.books)
    
    def __repr__(self):
        return f"Bookcollection containing ({self.books})"

    def __str__(self):
        return f"BookCollection with {len(self)} books, being ({self.books})"

    def __getitem__(self, key):
        return self.books[key]

    def __add__(self, other):
        return BookCollection(self.books + other.books)
    
    def __gt__(self, other):
        return len(self.books) > len(other.books)
```

```python
collection1 = BookCollection(["1984", "Brave New World"]) # kalder __init__
collection2 = BookCollection(["Dune"])                    # kalder __init__
collection3 = BookCollection([])                          # kalder __init__


print(collection1[0])                                     # kalder __getitem__
print(collection1)                                        # kalder __str__
print(collection3)                                        # kalder __str__

combined = collection1 + collection2 + collection3        # kalder __add__
print(combined)                                           # kalder __str__

print(collection1 > collection2)                          # kalder __gt__
```
